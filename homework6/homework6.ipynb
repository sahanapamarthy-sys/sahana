```python
matrix = np.array([
[5, 12, 17, 9],
[8, 21, 14, 3],
[6, 13, 19, 11]
])
```

```python
print("Overall mean:", np.mean(matrix))
```

```python
print("Max of each row:", np.max(matrix, axis=1))
print("Min of each column:", np.min(matrix, axis=0))
```

```python
flatten_array= matrix.flatten()
print("Flattened array:", flatten_array)
print("New shape:", flatten_array.shape)
```

```python
min_num = np.argmin(matrix)
max_num = np.argmax(matrix)
print("Flat index of min:", min_num)
print("Flat index of max:", max_num)
```

```python
min_coords = np.unravel_index(min_num, matrix.shape)
max_coords = np.unravel_index(max_num, matrix.shape)
print("Coordinates of min:", min_coords)
print("Coordinates of max:", max_coords)
```

```python
greater_15 = matrix[matrix > 15]
print("Values > 15:", greater_15)
```

```python
sorted_matrix = np.sort(matrix, axis=1)
print("Row-wise sorted matrix:\n", sorted_matrix)
```

```python
arr_second = np.array([
[2, 3, 5],
[14, 6, 8],
[11, 13, 17],
[7, 10, 13]
])

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
          return False
    return True
def rows_with_prime(arr_second):
    return arr_second[np.any(np.vectorize(is_prime)(arr_second), axis=1)]
print(rows_with_prime(arr_second))
```

```python
def checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[::2, ::2] = 1
    board[1::2, 1::2] = 1
    return board
print(checkerboard())

def reverse_checkerboard():
    board = np.zeros((8, 8), dtype=int)
    board[::2, 1::2] = 1
    board[1::2, ::2] = 1
    return board
print(reverse_checkerboard())
```

```python
arr = np.array(["galaxy", "clusters"])
def expansion(arr, n):
    spaces = " " * n
    return np.array([spaces.join(word) for word in arr])
print(expansion(arr, 3))
```
